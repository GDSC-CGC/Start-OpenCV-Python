```python
import cv2
import matplotlib.pyplot as plt
```

```python
font                   = cv2.FONT_HERSHEY_SIMPLEX
bottom                 = (200,300)
fontScale              = 1
fontColor              = (255,255,255)
lineType               = 2
```

```python
image = cv2.imread("dragon.png",0) ### OPENING AS GRAYSCALE...
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.putText(gray_img,'DROGON', 
    bottom, 
    font, 
    fontScale,
    fontColor,
    lineType)
plt.imshow(gray_img)

```

```python
image = cv2.imread("dragon.png")  ### DEFAULT BGR FORMAT
cv2.putText(image,'DROGON_IN_BGR', 
    bottom, 
    font, 
    fontScale,
    fontColor,
    lineType)
plt.imshow(image)
```

```python
image = cv2.imread("dragon.png")         ## BGR TP RGB CONVERSION
rgb_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.putText(rgb_image,'DROGON_IN_RGB', 
    bottom, 
    font, 
    fontScale,
    fontColor,
    lineType)
plt.imshow(rgb_image)
```
